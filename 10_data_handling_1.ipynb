```python
import pandas as pd
import numpy as np
```

# concat()
### pd.concat(
    objs,
    axis=0,
    join='outer',
    ignore_index: bool = False,
    keys=None,
    levels=None,
    names=None,
    verify_integrity: bool = False,
    sort: bool = False,
    copy: bool = True
)

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    )
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    )
df2
```

```python
result = pd.concat([df1,df2])
result
```

# concat()
### parameter : ignore_index
- ignore_index : bool, default False
    * False : 연결되기 전의 index 값을 그대로 사용
    * True : 새로운 index 값을 할당 (0, 1, ..., n-1)

```python
result = pd.concat([df1,df2], ignore_index= True)
result
```

# concat()
### parameter : axis
- axis : {0/'index', 1/'columns'}, default 0
   * 0 : 행(index)간 연결
   * 1 : 열(columns)간 연결

```python
df1
```

```python
df3 = pd.DataFrame({'E': ['E0', 'E1', 'E2', 'E3'],
                    'F': ['F0', 'F1', 'F2', 'F3'],},
                    )
df3
```

```python
result = pd.concat([df1,df3], axis= 1)
result
```

# concat()
### parameter : join
- join : {'inner', 'outer'}, default 'outer'
    * outer : 데이터 손실이 없는 결합연결(union)
    * inner : 공통된 부분만 교차연결(intersection)

```python
df1
```

```python
df4 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],},
                    )
df4
```

```python
result = pd.concat([df1,df4])
result
```

```python
result = pd.concat([df1,df4], join= 'inner', ignore_index= True)
result
```

# df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
Docstring:
Remove missing values.

Parameters
----------
- axis : {0 or 'index', 1 or 'columns'}, default 0
    * 0, or 'index' : 결측치를 가진 행 제거.
    * 1, or 'columns' : 결측치를 가진 열 제거.

- how : {'any', 'all'}, default 'any'
    * 'any' : 하나라도 NA 값을 가지면 해당 행 또는 열 제거.
    * 'all' : 모든값이 NA 이면 해당 행 또는 열 제거.
- inplace : bool, default False
    * True : 기존 DataFrame에 덮어쓰기.

```python
result = pd.concat([df1,df4])
result.dropna(axis=1, inplace=True)
result
```

```python
result = pd.concat([df1,df4])
result = result.dropna(axis=1)
result
```

# df.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None) 

Parameters
----------
- method : {'backfill', 'bfill', 'pad', 'ffill', None}, default None
    * pad / ffill: 앞의 유효한 값으로 채우기
    * backfill / bfill: 뒤의 유효한 값으로 채우기
- limit : int, default None
    * NaN값을 채우기할 최대 수 지정

```python
result = pd.concat([df1,df4])
result = result.fillna(100)
result
```

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                   columns=list('ABCD'))
df
```

```python
result = df.dropna(axis=1, how='all')
result
```

```python
result = df.fillna(method='ffill')
result
```

```python
result = df.fillna({'A': 0, 'B': 1, 'C': 2, 'D': 3})
result
```

```python
result = df.fillna({'A': 0, 'B': 1, 'C': 2, 'D': 3}, limit=1)
result
```

- keys : sequence, default None
    * 전달된 keys 값들을 이용해 가장 바깥쪽 레벨의 계층적 index를 생성
